```python
#to check missing values
df.isnull().sum()
```

```python
# to replace the null values
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x :x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#to change the column names 
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# To create a new column age_group
labels = ['Young' , 'Adult' , 'Middle Age' , 'Senior']
df['age_group'] = pd.qcut(df['age'] , q = 4 , labels = labels)
```

```python
df[['age' , 'age_group']]
```

```python
# to create a new column purchase_frequency_days
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-weekly':14,
    'Annually':365,
    'Every 3 months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
#check two columns are equal if equal remove one column
df[['discount_applied', 'promo_code_used']].head(10)
```
